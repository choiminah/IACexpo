```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/drive/MyDrive/movie/movies.csv')
ratings = pd.read_csv('/content/drive/MyDrive/movie/ratings.csv')
```

```python
movies.head()
```

```python
ratings.drop('timestamp', axis=1, inplace=True)
```

```python
ratings.head()
```

```python
df = pd.merge(ratings, movies, on='movieId')
df.head()
```

```python
mudf = df.pivot_table(index='title', columns='userId', values='rating')
```

```python
mudf.head()
```

```python
mudf.fillna(0, inplace=True)
mudf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 코사인 유사도 계산
cosine_similarity(mudf)
```

```python
itembasedcollaborative = pd.DataFrame(data=cosine_similarity(mudf), index=mudf.index, columns=mudf.index)
```

```python
def get_item_based_collaborative(title):
  return itembasedcollaborative[title].sort_values(ascending=False)[:6]
```

```python
get_item_based_collaborative('Avengers: Age of Ultron (2015)')
```
